```python
df1=df.drop(columns=['Unnamed: 32','id'],axis=1)
```

```python
x=df1.drop('diagnosis',axis=1)
y=df1.diagnosis
```

```python
x
```

```python
y1=y.replace({'M':0,'B':1})
y1
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

x_train,x_test,y_train,y_test=train_test_split(x,y1,test_size=0.2,stratify=y)

x_train
y_train
```

```python
type(x_train)
```

```python
type(y_train)
```

```python
scaler=StandardScaler()
x_train_std=pd.DataFrame(scaler.fit_transform(x_train))
x_test_std=pd.DataFrame(scaler.fit_transform(x_test))
```

```python
x_train_std
```

```python
x_test_std
```

# build model

```python
from sklearn.ensemble import RandomForestClassifier

clf=RandomForestClassifier(criterion="gini",max_depth=8,min_samples_split=10,random_state=5)

```

```python
clf.fit(x_train_std,y_train)
```

```python
y_pred=clf.predict(x_test_std)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score

print ("Accuracy score = ",accuracy_score(y_test,y_pred))
#print("Mean Absolute Error = ",MeanAbsoluteError(y_test,y_pred))
```

```python
from sklearn.model_selection import cross_val_score

print("Cross Validation Score = ",cross_val_score(clf,x_train_std,y_train,cv=10))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred,y_test))
```
